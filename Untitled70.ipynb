```python
import numpy as np

all_tags = set()
for ex in ds:
    all_tags.update(ex["ner_tags"])
label_list = sorted(all_tags)
label2id = {l: i for i, l in enumerate(label_list)}
id2label = {i: l for l, i in label2id.items()}

print("Метки:", label_list)
print("Кол-во меток:", len(label_list))

ds_split = ds.train_test_split(test_size=0.2, seed=42)
val_test = ds_split["test"].train_test_split(test_size=0.5, seed=42)
train_ds = ds_split["train"]
val_ds   = val_test["train"]
test_ds  = val_test["test"]

print(f"Train: {len(train_ds)}, Val: {len(val_ds)}, Test: {len(test_ds)}")
```

```python
from transformers import AutoTokenizer

MODEL_NAME = "bert-base-multilingual-cased"

tokenizer  = AutoTokenizer.from_pretrained(MODEL_NAME)

def tokenize_and_align(examples):
    tokenized = tokenizer(
        examples["tokens"],
        truncation=True,
        is_split_into_words=True,
        padding="max_length",
        max_length=256
    )
    all_labels = []
    for i, labels in enumerate(examples["ner_tags"]):
        word_ids  = tokenized.word_ids(batch_index=i)
        label_ids = []
        prev_word = None
        for wid in word_ids:
            if wid is None:
                label_ids.append(-100)
            elif wid != prev_word:
                label_ids.append(label2id[labels[wid]])
            else:
                lbl = labels[wid]
                label_ids.append(label2id[lbl.replace("B-", "I-") if lbl.startswith("B-") else lbl])
            prev_word = wid
        all_labels.append(label_ids)
    tokenized["labels"] = all_labels
    return tokenized

train_tok = train_ds.map(tokenize_and_align, batched=True, remove_columns=train_ds.column_names)
val_tok   = val_ds.map(tokenize_and_align,   batched=True, remove_columns=val_ds.column_names)
test_tok  = test_ds.map(tokenize_and_align,  batched=True, remove_columns=test_ds.column_names)

train_tok.set_format("torch")
val_tok.set_format("torch")
test_tok.set_format("torch")

print("✅ Токенизация готова")
```

```python
from seqeval.metrics import f1_score, precision_score, recall_score
import torch

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Устройство:", DEVICE)

def compute_metrics(predictions, labels):
    preds = np.argmax(predictions, axis=2)
    true_labels, true_preds = [], []
    for pred_seq, label_seq in zip(preds, labels):
        tl, tp = [], []
        for p, l in zip(pred_seq, label_seq):
            if l != -100:
                tl.append(id2label[l])
                tp.append(id2label[p])
        true_labels.append(tl)
        true_preds.append(tp)
    return {
        "f1":        f1_score(true_labels, true_preds),
        "precision": precision_score(true_labels, true_preds),
        "recall":    recall_score(true_labels, true_preds),
    }

def evaluate_model(model, dataset, device):
    model.to(device)
    model.eval()
    loader = torch.utils.data.DataLoader(dataset, batch_size=16)
    all_preds, all_labels = [], []
    with torch.no_grad():
        for batch in loader:
            outputs = model(
                input_ids=batch["input_ids"].to(device),
                attention_mask=batch["attention_mask"].to(device)
            )
            all_preds.append(outputs.logits.cpu().numpy())
            all_labels.append(batch["labels"].numpy())
    return compute_metrics(
        np.concatenate(all_preds,  axis=0),
        np.concatenate(all_labels, axis=0)
    )

print("✅ Функции метрик готовы")
```

```python
from transformers import get_scheduler
from torch.optim import AdamW
import torch, numpy as np

def train_model(model, train_dataset, val_dataset,
                epochs=3, lr=2e-5, print_every=20):
    model.to(DEVICE)
    loader = torch.utils.data.DataLoader(train_dataset, batch_size=16, shuffle=True)

    optimizer = AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=lr)

    total_steps = len(loader) * epochs
    scheduler = get_scheduler(
        "linear",
        optimizer=optimizer,
        num_warmup_steps=int(0.1 * total_steps),   # 10% шагов на warmup
        num_training_steps=total_steps,
    )

    for epoch in range(epochs):
        model.train()
        total_loss = 0.0

        for step, batch in enumerate(loader):
            optimizer.zero_grad()

            outputs = model(
                input_ids=batch["input_ids"].to(DEVICE),
                attention_mask=batch["attention_mask"].to(DEVICE),
                labels=batch["labels"].to(DEVICE),
            )
            loss = outputs.loss
            loss.backward()
            optimizer.step()
            scheduler.step()

            total_loss += loss.item()

            # печать лосса каждые print_every батчей
            if (step + 1) % print_every == 0:
                avg_loss = total_loss / (step + 1)
                print(f"Epoch {epoch+1} | step {step+1}/{len(loader)} | loss={avg_loss:.4f}")

        # в конце эпохи — валидация
        val_m = evaluate_model(model, val_dataset, DEVICE)
        print(f"Epoch {epoch+1} DONE | train_loss={total_loss/len(loader):.4f} "
              f"| val F1={val_m['f1']:.4f} P={val_m['precision']:.4f} R={val_m['recall']:.4f}")

    return model
```

```python
from transformers import AutoModelForTokenClassification

print("=" * 50)
print("BASELINE — BioBERT без файнтюна")
print("=" * 50)

model_base = AutoModelForTokenClassification.from_pretrained(
    MODEL_NAME, num_labels=len(label_list),
    id2label=id2label, label2id=label2id,
    ignore_mismatched_sizes=True
)
model_base.to(DEVICE)
baseline_metrics = evaluate_model(model_base, test_tok, DEVICE)
print(f"F1={baseline_metrics['f1']:.4f} | P={baseline_metrics['precision']:.4f} | R={baseline_metrics['recall']:.4f}")
```

```python
print("=" * 50)
print("S1 — Freeze bottom-6 layers (DeBERTa)")
print("=" * 50)

model_s1 = AutoModelForTokenClassification.from_pretrained(
    MODEL_NAME,
    num_labels=len(label_list),
    id2label=id2label,
    label2id=label2id,
    ignore_mismatched_sizes=True,
)

# Замораживаем embeddings
for param in model_s1.deberta.embeddings.parameters():
    param.requires_grad = False

# Замораживаем первые 6 слоёв энкодера
for i, layer in enumerate(model_s1.deberta.encoder.layer):
    if i < 6:
        for param in layer.parameters():
            param.requires_grad = False

print(f"Обучаемых параметров: {sum(p.numel() for p in model_s1.parameters() if p.requires_grad):,}")
model_s1 = train_model(model_s1, train_tok, val_tok)
s1_metrics = evaluate_model(model_s1, test_tok, DEVICE)
print(f"✅ S1 → F1={s1_metrics['f1']:.4f} | P={s1_metrics['precision']:.4f} | R={s1_metrics['recall']:.4f}")
```

```python
print("=" * 50)
print("S2 — Tokenizer only")
print("=" * 50)

tokenizer_s2 = AutoTokenizer.from_pretrained(MODEL_NAME)
chem_vocab = [
    "SMILES", "IUPAC", "Ki", "IC50", "EC50", "Kd", "microM", "nM", "pM",
    "NMDA", "sigma-1", "sigma-2", "phosphate", "morphinan", "dextrorphan",
    "anticonvulsant", "neuroprotective", "CPR-3", "CPR-4",
    "agonist", "antagonist", "inhibitor", "substrate"
]
added = tokenizer_s2.add_tokens(chem_vocab)
print(f"Добавлено токенов: {added}")

model_s2 = AutoModelForTokenClassification.from_pretrained(
    MODEL_NAME, num_labels=len(label_list),
    id2label=id2label, label2id=label2id,
    ignore_mismatched_sizes=True
)
model_s2.resize_token_embeddings(len(tokenizer_s2))
for name, param in model_s2.named_parameters():
    if "embeddings" not in name:
        param.requires_grad = False

def tokenize_and_align_s2(examples):
    tokenized = tokenizer_s2(
        examples["tokens"], truncation=True,
        is_split_into_words=True, padding="max_length", max_length=256
    )
    all_labels = []
    for i, labels in enumerate(examples["ner_tags"]):
        word_ids  = tokenized.word_ids(batch_index=i)
        label_ids, prev_word = [], None
        for wid in word_ids:
            if wid is None:
                label_ids.append(-100)
            elif wid != prev_word:
                label_ids.append(label2id[labels[wid]])
            else:
                lbl = labels[wid]
                label_ids.append(label2id[lbl.replace("B-", "I-") if lbl.startswith("B-") else lbl])
            prev_word = wid
        all_labels.append(label_ids)
    tokenized["labels"] = all_labels
    return tokenized

train_tok_s2 = train_ds.map(tokenize_and_align_s2, batched=True, remove_columns=train_ds.column_names)
val_tok_s2   = val_ds.map(tokenize_and_align_s2,   batched=True, remove_columns=val_ds.column_names)
test_tok_s2  = test_ds.map(tokenize_and_align_s2,  batched=True, remove_columns=test_ds.column_names)
train_tok_s2.set_format("torch"); val_tok_s2.set_format("torch"); test_tok_s2.set_format("torch")

model_s2   = train_model(model_s2, train_tok_s2, val_tok_s2)
s2_metrics = evaluate_model(model_s2, test_tok_s2, DEVICE)
print(f"✅ S2 → F1={s2_metrics['f1']:.4f} | P={s2_metrics['precision']:.4f} | R={s2_metrics['recall']:.4f}")
```

```python
print("=" * 50)
print("S3 — Full fine-tuning")
print("=" * 50)

model_s3 = AutoModelForTokenClassification.from_pretrained(
    MODEL_NAME, num_labels=len(label_list),
    id2label=id2label, label2id=label2id,
    ignore_mismatched_sizes=True
)
print(f"Обучаемых параметров: {sum(p.numel() for p in model_s3.parameters() if p.requires_grad):,}")
model_s3   = train_model(model_s3, train_tok, val_tok)
s3_metrics = evaluate_model(model_s3, test_tok, DEVICE)
print(f"✅ S3 → F1={s3_metrics['f1']:.4f} | P={s3_metrics['precision']:.4f} | R={s3_metrics['recall']:.4f}")
```
