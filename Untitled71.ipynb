```python
import os, re, numpy as np, torch
from datasets import load_from_disk, Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForTokenClassification,
    AutoModelForMaskedLM,
    DataCollatorForLanguageModeling,
    Trainer,
    TrainingArguments,
    get_scheduler,
)

```

```python
from seqeval.metrics import f1_score, precision_score, recall_score
from torch.optim import AdamW

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", DEVICE)
```

```python
# Путь к сохранённому датасету
from google.colab import files
uploaded = files.upload()  # загружаешь chem_hf_dataset.zip

!unzip chem_hf_dataset.zip -d /tmp/extracted/
!mv /tmp/extracted/content/chem_hf_dataset /content/chem_hf_dataset

from datasets import load_from_disk
ds = load_from_disk("/content/chem_hf_dataset")
print(ds)

# Собираем список меток
all_tags = set()
for ex in ds:
    all_tags.update(ex["ner_tags"])
label_list = sorted(all_tags)
label2id = {l: i for i, l in enumerate(label_list)}
id2label = {i: l for l, i in label2id.items()}

print("Метки:", label_list)
print("Кол-во меток:", len(label_list))

# Сплит 80/10/10
ds_split = ds.train_test_split(test_size=0.2, seed=42)
val_test = ds_split["test"].train_test_split(test_size=0.5, seed=42)

train_ds = ds_split["train"]
val_ds   = val_test["train"]
test_ds  = val_test["test"]

print(f"Train: {len(train_ds)}, Val: {len(val_ds)}, Test: {len(test_ds)}")
```

```python
def make_tokenize_and_align(tokenizer, max_length=256):
    def tokenize_and_align(examples):
        tokenized = tokenizer(
            examples["tokens"],
            truncation=True,
            is_split_into_words=True,
            padding="max_length",
            max_length=max_length,
        )
        all_labels = []
        for i, labels in enumerate(examples["ner_tags"]):
            word_ids  = tokenized.word_ids(batch_index=i)
            label_ids = []
            prev_word = None
            for wid in word_ids:
                if wid is None:
                    label_ids.append(-100)
                elif wid != prev_word:
                    label_ids.append(label2id[labels[wid]])
                else:
                    lbl = labels[wid]
                    label_ids.append(label2id[lbl.replace("B-", "I-") if lbl.startswith("B-") else lbl])
                prev_word = wid
            all_labels.append(label_ids)
        tokenized["labels"] = all_labels
        return tokenized
    return tokenize_and_align
```

```python
def compute_metrics(predictions, labels):
    preds = np.argmax(predictions, axis=2)
    true_labels, true_preds = [], []
    for pred_seq, label_seq in zip(preds, labels):
        tl, tp = [], []
        for p, l in zip(pred_seq, label_seq):
            if l != -100:
                tl.append(id2label[l])
                tp.append(id2label[p])
        true_labels.append(tl)
        true_preds.append(tp)
    return {
        "f1":        f1_score(true_labels, true_preds),
        "precision": precision_score(true_labels, true_preds),
        "recall":    recall_score(true_labels, true_preds),
    }

def evaluate_model(model, dataset, device=DEVICE, batch_size=16):
    model.to(device)
    model.eval()
    loader = torch.utils.data.DataLoader(dataset, batch_size=batch_size)
    all_preds, all_labels = [], []
    with torch.no_grad():
        for batch in loader:
            outputs = model(
                input_ids=batch["input_ids"].to(device),
                attention_mask=batch["attention_mask"].to(device),
            )
            all_preds.append(outputs.logits.cpu().numpy())
            all_labels.append(batch["labels"].numpy())
    all_preds  = np.concatenate(all_preds,  axis=0)
    all_labels = np.concatenate(all_labels, axis=0)
    return compute_metrics(all_preds, all_labels)
```

```python
def train_model(model, train_dataset, val_dataset,
                epochs=3, lr=2e-5, print_every=20, batch_size=16):
    model.to(DEVICE)
    loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

    optimizer = AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=lr)

    total_steps = len(loader) * epochs
    scheduler = get_scheduler(
        "linear",
        optimizer=optimizer,
        num_warmup_steps=int(0.1 * total_steps),   # 10% warmup
        num_training_steps=total_steps,
    )

    for epoch in range(epochs):
        model.train()
        total_loss = 0.0

        for step, batch in enumerate(loader):
            optimizer.zero_grad()
            outputs = model(
                input_ids=batch["input_ids"].to(DEVICE),
                attention_mask=batch["attention_mask"].to(DEVICE),
                labels=batch["labels"].to(DEVICE),
            )
            loss = outputs.loss
            loss.backward()
            optimizer.step()
            scheduler.step()

            total_loss += loss.item()
            if (step + 1) % print_every == 0:
                avg_loss = total_loss / (step + 1)
                print(f"Epoch {epoch+1} | step {step+1}/{len(loader)} | loss={avg_loss:.4f}")

        val_m = evaluate_model(model, val_dataset, DEVICE, batch_size)
        print(f"Epoch {epoch+1} DONE | train_loss={total_loss/len(loader):.4f} "
              f"| val F1={val_m['f1']:.4f} P={val_m['precision']:.4f} R={val_m['recall']:.4f}")

    return model
```

Часть A. mBERT до DAPT (baseline)
Cell A1 — Настройка mBERT и токенизация

```python
MODEL_NAME_MBERT = "bert-base-multilingual-cased"

tokenizer_mbert = AutoTokenizer.from_pretrained(MODEL_NAME_MBERT)
tokenize_and_align_mbert = make_tokenize_and_align(tokenizer_mbert, max_length=256)

train_tok_mbert = train_ds.map(tokenize_and_align_mbert, batched=True, remove_columns=train_ds.column_names)
val_tok_mbert   = val_ds.map(tokenize_and_align_mbert,   batched=True, remove_columns=val_ds.column_names)
test_tok_mbert  = test_ds.map(tokenize_and_align_mbert,  batched=True, remove_columns=test_ds.column_names)

train_tok_mbert.set_format("torch")
val_tok_mbert.set_format("torch")
test_tok_mbert.set_format("torch")
print("mBERT токенизация готова")
```

Cell A2 — Baseline + S1 + S2 + S3 для mBERT

```python
print("=" * 50)
print("mBERT — Baseline")
print("=" * 50)

model_base_mbert = AutoModelForTokenClassification.from_pretrained(
    MODEL_NAME_MBERT,
    num_labels=len(label_list),
    id2label=id2label,
    label2id=label2id,
    ignore_mismatched_sizes=True,
)
baseline_mbert = evaluate_model(model_base_mbert, test_tok_mbert)
print(baseline_mbert)
```

```python
print("=" * 50)
print("mBERT — S1 Freeze bottom-6")
print("=" * 50)

model_s1_mbert = AutoModelForTokenClassification.from_pretrained(
    MODEL_NAME_MBERT,
    num_labels=len(label_list),
    id2label=id2label,
    label2id=label2id,
    ignore_mismatched_sizes=True,
)

for p in model_s1_mbert.bert.embeddings.parameters():
    p.requires_grad = False
for i, layer in enumerate(model_s1_mbert.bert.encoder.layer):
    if i < 6:
        for p in layer.parameters():
            p.requires_grad = False

print("Trainable params S1:", sum(p.numel() for p in model_s1_mbert.parameters() if p.requires_grad))
model_s1_mbert = train_model(model_s1_mbert, train_tok_mbert, val_tok_mbert, epochs=3, lr=2e-5)
s1_mbert = evaluate_model(model_s1_mbert, test_tok_mbert)
print(s1_mbert)
```

```python
print("=" * 50)
print("mBERT — S2 Tokenizer only")
print("=" * 50)

tokenizer_s2_mbert = AutoTokenizer.from_pretrained(MODEL_NAME_MBERT)
chem_vocab = [
    "SMILES", "IUPAC", "Ki", "IC50", "EC50", "Kd", "microM", "nM", "pM",
    "NMDA", "sigma-1", "sigma-2", "phosphate", "morphinan", "dextrorphan",
    "anticonvulsant", "neuroprotective", "CPR-3", "CPR-4",
    "agonist", "antagonist", "inhibitor", "substrate"
]
added = tokenizer_s2_mbert.add_tokens(chem_vocab)
print("Добавлено токенов:", added)

model_s2_mbert = AutoModelForTokenClassification.from_pretrained(
    MODEL_NAME_MBERT,
    num_labels=len(label_list),
    id2label=id2label,
    label2id=label2id,
    ignore_mismatched_sizes=True,
)
model_s2_mbert.resize_token_embeddings(len(tokenizer_s2_mbert))

for name, p in model_s2_mbert.named_parameters():
    if "embeddings" not in name:
        p.requires_grad = False

tokenize_and_align_s2_mbert = make_tokenize_and_align(tokenizer_s2_mbert, max_length=256)
train_tok_s2m = train_ds.map(tokenize_and_align_s2_mbert, batched=True, remove_columns=train_ds.column_names)
val_tok_s2m   = val_ds.map(tokenize_and_align_s2_mbert,   batched=True, remove_columns=val_ds.column_names)
test_tok_s2m  = test_ds.map(tokenize_and_align_s2_mbert,  batched=True, remove_columns=test_ds.column_names)
train_tok_s2m.set_format("torch"); val_tok_s2m.set_format("torch"); test_tok_s2m.set_format("torch")

model_s2_mbert = train_model(model_s2_mbert, train_tok_s2m, val_tok_s2m, epochs=3, lr=2e-5)
s2_mbert = evaluate_model(model_s2_mbert, test_tok_s2m)
print(s2_mbert)
```

```python
print("=" * 50)
print("mBERT — S3 Full fine-tuning")
print("=" * 50)

model_s3_mbert = AutoModelForTokenClassification.from_pretrained(
    MODEL_NAME_MBERT,
    num_labels=len(label_list),
    id2label=id2label,
    label2id=label2id,
    ignore_mismatched_sizes=True,
)
print("Trainable params S3:", sum(p.numel() for p in model_s3_mbert.parameters() if p.requires_grad))
model_s3_mbert = train_model(model_s3_mbert, train_tok_mbert, val_tok_mbert, epochs=3, lr=2e-5)
s3_mbert = evaluate_model(model_s3_mbert, test_tok_mbert)
print(s3_mbert)
```

Часть B. DAPT для mBERT (MLM‑предобучение)
Cell B1 — Подготовка корпуса MLM

```python
# Строим корпус текстов для MLM из тех же документов
mlm_ds = ds.map(
    lambda ex: {"text": " ".join(ex["tokens"])},
    remove_columns=ds.column_names
)
print(mlm_ds[0]["text"])
```

```python
tokenizer_mlm = AutoTokenizer.from_pretrained(MODEL_NAME_MBERT)

def tokenize_mlm(examples):
    return tokenizer_mlm(
        examples["text"],
        truncation=True,
        padding="max_length",
        max_length=256,
    )

tokenized_mlm = mlm_ds.map(tokenize_mlm, batched=True, remove_columns=["text"])
tokenized_mlm.set_format("torch")

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer_mlm,
    mlm=True,
    mlm_probability=0.15,
)

mlm_model = AutoModelForMaskedLM.from_pretrained(MODEL_NAME_MBERT)

training_args = TrainingArguments(
    output_dir="mbert_dapt_chem",
    num_train_epochs=3,
    per_device_train_batch_size=16,
    learning_rate=5e-5,
    weight_decay=0.01,
    logging_steps=50,
    save_steps=500,
    report_to="none",
)

trainer = Trainer(
    model=mlm_model,
    args=training_args,
    train_dataset=tokenized_mlm,
    data_collator=data_collator,
)

trainer.train()

mlm_model.save_pretrained("mbert_dapt_chem")
tokenizer_mlm.save_pretrained("mbert_dapt_chem")
print("✅ DAPT для mBERT завершён, чекпойнт mbert_dapt_chem сохранён")
```
